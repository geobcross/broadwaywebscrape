```
import requests 
import pandas as pd
from bs4 import BeautifulSoup
```

```
url3 = 'https://www.ibdb.com/shows/'
response3 = requests.get(url3)
soup3 = BeautifulSoup(response3.text, 'lxml')
soup3.find('div')
all_divs = soup3.find_all('div')
current = soup3.find_all('div',{'class': "xt-c-box row"})
current = soup3.find_all('div',{'class': "xt-iblock-inner"})

```

```
anchors = soup3.findAll('a')






```

```
current_links = anchors[11:38]
current_links
```

```

now_showing = []
base_url = 'https://www.ibdb.com'
now_showing.append([base_url + a.attrs['href'] for a in current_links])

now_showing_urls = now_showing
now_showing_urls = now_showing_urls[0]
now_showing_urls
```

```

attempt = soup3.find_all('i')
attempt = list(attempt)
```

```

on_broadway = attempt[18:45]
on_broadway
```

```
len(on_broadway)
```

```

currently_on_broadway = []
for i in on_broadway:
  currently_on_broadway.append(i.text)


currently_on_broadway
```

```
openings = []
for i in now_showing_urls:
  response2 = requests.get(i)
  soup2 = BeautifulSoup(response2.text, 'lxml')
  opening = soup2.find('div', {'class':"col s5 m3 l5 txt-paddings"})
  openings.append(opening.find('div',{'class':"xt-main-title"}))

openings 
```

```
opening_night = []
for i in openings:
  opening_night.append(i.text)
```

```
opening_night
```

```
CURRENT_SHOWS = {'Title': currently_on_broadway, 'opening_night': opening_night}
```

```
my_df = pd.DataFrame(CURRENT_SHOWS)
```
