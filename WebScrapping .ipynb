```
url = 'https://pitchfork.com/reviews/albums/'

#First home page 
response = requests.get(url)
soup = BeautifulSoup(response.text, 'lxml')
rev_1 = soup.find('div', {'class':'review'})
#artist
artist = rev_1.find('div', {'class':'review__title'}).find('li').text
#title
title = rev_1.find('div', {'class':'review__title'}).find('h2').text


```

```
## Go througyh each url and do 
full_review_url = rev_1.find('a').attrs['href']
base_url = 'https://www.pitchfork.com'
full_review_url = base_url + full_review_url
response2 = requests.get(full_review_url)
soup2 = BeautifulSoup(response2.text, 'lxml')
score1 = soup2.find('span', {'class':'score'}).text
score1

```

```
#all info
all_info = soup.find_all('div', {'class':'review'})
#The artist names 
artists = [i.find('li').text for i in all_info]
titles = [i.find('h2').text for i in all_info]
full_rev_urls = ['https://www.pitchfork.com'+ i.find('a').attrs['href'] for i in all_info]
```

```
all_info
```

```
artists
```

```
titles
```

```
full_rev_urls 
```

```
scores = []
for i in full_rev_urls:
  response2 = requests.get(i)
  soup2 = BeautifulSoup(response2.text, 'lxml')
  scores.append(soup2.find('span', {'class':'score'}))
  scores
  
```

```
scores2= []
for i in scores:
  scores2.append(i.text)
```

```
scores2
```

```
# title for all on page 1 

#artists for all on page 1 

#scores for all on page 1 
```

```
import pandas as pd

pitchfork = pd.DataFrame({'artist': artists, 'title': titles, 'score':scores2})
```
